```python
import vqe_methods
import operator_pools
```

### singlet_GSD pool

```python
r = 1.0
geometry = [('H', (0,0,1*r)), ('H', (0,0,2*r)), ('H', (0,0,3*r)), ('H', (0,0,4*r))]
vqe_methods.adapt_vqe(geometry, pool = operator_pools.singlet_GSD())
```

## singlet_GSD with [0,2,4,6] reference_ket (expected to fail)

```python
vqe_methods.adapt_vqe(geometry, ref_state = [0,2,4,6], pool = operator_pools.singlet_GSD())
```

## singlet_GSD with spin multiplicity 5 and [0,2,4,6] reference ket (also expected to fail)

```python
vqe_methods.adapt_vqe(geometry, multiplicity = 5, reference = 'rohf', ref_state =[0,2,4,6], pool = operator_pools.singlet_GSD())
```

## spin_flip_high_spin_rohf_GSD
This pool does not spin adapt and let the indices run over all spin orbitals, so it allows for the excitations already in the GSD pool and also spin-flips.

```python
vqe_methods.adapt_vqe(geometry, multiplicity = 5, reference = 'rohf', ref_state =[0,2,4,6], pool = operator_pools.sf_hs_rohf_GSD())
```

## spin_flip_high_spin_rohf_GSD with a HF state

```python
vqe_methods.adapt_vqe(geometry, pool = operator_pools.sf_hs_rohf_GSD())
```
